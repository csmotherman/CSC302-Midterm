```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Sleep_health_and_lifestyle_dataset.csv')

sns.boxplot(data=df, x='Sleep Disorder', y='Quality of Sleep', palette=sns.husl_palette(h=.5))
```

```python
sns.boxplot(data=df, x='Sleep Disorder', y='Sleep Duration', palette=sns.husl_palette(h=.5))
```

```python
sns.displot(data=df, x="Stress Level", hue="Sleep Disorder", fill=True, multiple="stack", kind="kde", palette="viridis")
plt.xlim(3, 8)
```
